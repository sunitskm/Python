```python
import matplotlib.pyplot as plt
import pandas
data = pandas.read_csv('ARVIND.NS.csv')
data["Date"] = pandas.DatetimeIndex(data["Date"])
data = data.set_index('Date')
data["Change"] = data["Open"]-data["Close"]
data['MA'] = data['Adj Close'].rolling(5).mean()
data.tail(10)

#data['Adj Close').plot()
data["Adj Close"].plot()
data["MA"].plot()
plt.legend()
plt.show()

#data
```
